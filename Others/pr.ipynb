```python
cuisines = list()
for recipe in a:
    cuisines.append(recipe['cuisine'])
print(len(cuisines))
```

```python
# number of recipes of each type, we can use this to count the ratio, which is used to split the data (cross-validation).
counts = dict()
for cuisine in cuisines:
    counts[cuisine] = counts.get(cuisine,0) + 1
print(counts)
print()
print(counts.keys())
print()
print(len(counts.keys()), 'receipes')
```

```python
# What ingredients occur in Greek receipe? Not 6000+, so there are many ingredients not in Greek cuisine. 
greek_ingrd = list()
s_greek = set()
for recipe in a:
    if recipe['cuisine'] == 'greek':
        greek_ingrd = greek_ingrd + recipe['ingredients']
        s_greek.update(recipe['ingredients'])
print('lenth_of_ingredients_without_duplicate:', len(s_greek))
print('lenth_of_ingredients_list_with_duplicate:', len(greek_ingrd))

print('Greek_ingredients_without_duplicate:', s_greek)

      
print('Greek_ingredients_list_with_duplicate:', greek_ingrd)

      
        
        
```

```python
#Counting how many times each ingredients occur in Greek receipe.
greek_counts = dict()
for s in greek_ingrd:
    greek_counts[s] = greek_counts.get(s,0) + 1
print(greek_counts)
```

```python
#Order ingredients and occurance number in decreasing order for Greek receipe.
tmp = list()
for k,v in greek_counts.items():
    tmp.append((v,k))
tmp = sorted(tmp, reverse = True)
print(tmp)
```
